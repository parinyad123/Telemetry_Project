### Import Data

```python
import pandas as pd

import calendar
import datetime
```

```python
# import data
data = pd.read_csv('TTH15A.csv',sep=",")
data
```

### Find Duplicate Rows based on all columns 

```python
# Select duplicate rows except first occurrence based on all columns

duplicateRowsDF = data[data.duplicated()]
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicateRowsDF)
```

### Find Duplicate Rows based on selected columns

```python
# Select all duplicate rows based on one column

duplicateRowsDF = data[data.duplicated(['generation_time'])]
print("Duplicate Rows based on a single column are:", duplicateRowsDF, sep='\n')
```

### Counts for multiindex

```python
# data.set_index(['generation_time', 'eng_value']).count(level='generation_time')

dups_color = data.pivot_table(index=['generation_time'], aggfunc='size')
print (dups_color)
```

```python
data.set_index(['generation_time', 'eng_value']).count(level='generation_time')
```

```python
dups_color_and_shape = data.pivot_table(index=['generation_time','eng_value'], aggfunc='size')
print (dups_color_and_shape)
```

```python
# sorting date of data by generation time 
# เรียงลำดับข้อมูลใหม่ตามเวลา
data.sort_values(by=['generation_time'], inplace=True)

data
```

```python
# ลบ generatioin_time = 0

# Get names of indexes for which column generation_time has value 0
indexNames = data[data['generation_time']==0].index

# Delete these row indexes from dataFrame
data.drop(indexNames, inplace=True)
data
```

```python
# remove duplicte values by generation_time  ลบค่าซ้ำ
# data.drop_duplicates(subset ="generation_time", keep = False, inplace = True)
data.drop_duplicates(inplace = True)
data
```

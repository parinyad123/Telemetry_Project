```python
name_col = "'MT'\n"QZ
into = "select table_name from information_schema.tables where table_schema = 'public' order by table_name;\n\n"

# open or create file
fh = open("codeSQL.txt", "w")

# write file
fh.write(into)
fh.write('select name, generation_time, eng_value from '+ df['table_name'][0]+ ' where name= '+name_col)

for i in range(len(df['table_name'])-1):
    j = i+1
    fh.write("UNION ALL\n")
    fh.write('select name, generation_time, eng_value from '+ df['table_name'][j]+ ' where name= '+name_col)
    
fh.write(";")

fh.close()
```
